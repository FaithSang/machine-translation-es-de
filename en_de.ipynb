```python
import gradio as gr    #UI Library
from gradio import Textbox, Interface
from transformers import pipeline   #Transformers pipeline
```

2 Load Up Pipeline

```python
translation_pipeline = pipeline('translation_en_to_de')
```

```python
results = translation_pipeline('I love ice cream')
```

```python
results[0]['translation_text']
```

3 Create Gradio Function and Interface

```python
def translate_transformers(from_text):
    results = translation_pipeline(from_text)
    return results[0]['translation_text']
```

```python
translate_transformers('My name is Faith Sang')
```

```python
#gradio.Interface= main class, next passed the function to be run, inputs what we want it to look like textbox, then output
interface = gr.Interface(fn=translate_transformers, 
                         inputs=[Textbox(label="Text to be translated")],
                         outputs='text')
```

```python
interface.launch()
```
